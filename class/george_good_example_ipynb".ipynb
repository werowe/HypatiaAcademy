````python
import numpy as np

#linespace

# Array of 20 odd numbers
odd_numbers = np.arange(start=1, stop=1 + 2 * 20, step=2)
print("20 odd numbers:", odd_numbers)

odd_numbers
````

````python
# Array of 20 even numbers
even_numbers = np.arange(0, 2 + 2 * 20, 2)
print("\n20 even numbers:", even_numbers)

# a scalar
scalar_array = np.array(7)
print("\nScalar:", scalar_array, "Shape:", scalar_array.shape)


````

````python
# a 1-D array (ndim=1)

# shape means the number of elements along each axes (dims)
array_1d = np.array([10, 20, 30, 40, 50])
print("\n1-D array:", array_1d, "Shape:", array_1d.shape, "ndim:", array_1d.ndim)
````
