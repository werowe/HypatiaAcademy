```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# cast means to coerce or convert from one data type another

x = np.linspace(1,300,10).astype(np.int16)

print(b'00000001')

y = 3*x

# multiply an array by a scala

# 2* x

# 2 [x1, x2, x3, ....]

# when you multiple two arrays that's called linear algebra



z = 4 * x + 100
print(x)

print("\n")

```

```python
# canvas the thing you draw on


# from the plt object I create a figure and an axis


# figure is like a canvas

# axes


fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(x,y, marker="o")
ax.plot(x,z, marker="s")

ax.legend()

ax.grid()

ax.set_title("X-Y line plot")

ax.set_xlabel("x",color="blue")
ax.set_ylabel("y")


plt.xticks(rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


arr=np.array([1,1,2,2,2,2,2,3,3,3,3,3,4,4,4,4])

# Create a histogram with bins for integers
bins = np.arange(arr.min(), arr.max() + 2)  # Add 2 to include the last integer bin

counts, bin_edges, _ = plt.hist(arr, bins=bins, align='mid', edgecolor='black')

# Set x-ticks at the center of each bar
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])  # Midpoints of bins

# Integer labels

plt.xticks(bin_centers, labels=np.unique(arr))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# getting the dataset
penguinsData = sns.load_dataset('penguins')

# creating a custom color palette to use in the barplot
# seaborn will now use this one as the default color palette for every barplot
customPalette = sns.color_palette(['#2a57de', '#cb32e3'])
sns.set_palette(customPalette)

# creating the barplot
sns.barplot(
    x = 'species',
    y = 'body_mass_g', # if we set the y axis to the mass parameter, it will automatically calculate the means for each sex in each species
    hue = 'sex', # telling the seaborn to divide all species' data into male and female columns
    data = penguinsData, # setting the dataset
    order = ['Adelie', 'Chinstrap', 'Gentoo'], # setting the order of species on the x axis
    capsize = 0.1, # size of caps that represent possible data deviations
    saturation = 0.9, # opacity of colors
    errorbar='sd' # setting the caps to show one std as a possible data deviation
)

# showing the barplot
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt


arr = np.random.randint(1,5,20)

print("arr sorted", np.sort(arr))

unique_values, counts = np.unique(arr, return_counts=True)

labels = np.unique(arr)

plt.pie(counts, labels=labels, colors=["yellow", "blue", "green", "red"], autopct="%1.1f%%", textprops={"color": "magenta"} )

plt.legend(title="Values")

plt.show()
```

```python
# add explode
import numpy as np

import matplotlib.pyplot as plt


unique_values, counts = np.unique(arr, return_counts=True)

labels = np.unique(arr)

explode=[0,0.1,0,0]

plt.pie(counts, labels=labels, explode=explode, colors=["yellow", "blue", "green", "red"], autopct="%1.1f%%")

plt.legend(title="Values")

plt.show()
```

```python
# this is data drawn from a normal distribution with mean 0 and standard deviation 3

import matplotlib.pyplot as plt

a = np.random.normal(loc=0, scale=3, size=25)

b= np.round(a).astype(int)
c= b.reshape(5,5)

plt.imshow(c, cmap='cool')

plt.colorbar()

for i in range(c.shape[0]):
    for j in range(c.shape[1]):
        plt.annotate(str(c[i, j]), xy=(j, i), ha='center', va='center', color='white')


plt.show()
```
