# Graphics

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Generate x values
x = np.linspace(0, 4 * np.pi, 1000)

# Compute sine and cosine values with varying amplitudes and frequencies
y_sin = np.sin(x)
y_cos = np.cos(x)
y_sin2 = 0.5 * np.sin(2 * x)
y_cos2 = 0.5 * np.cos(2 * x)

# Generate random scatter data
np.random.seed(42)
x_scatter = np.random.uniform(0, 4 * np.pi, 300)
y_scatter = np.random.uniform(-1, 1, 300)

# Create a custom colormap for scatter points
cmap = ListedColormap(['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])

# Create the plot
plt.figure(figsize=(12, 8))

# Plot sine and cosine waves with different colors and styles
plt.plot(x, y_sin, label='sin(x)', color='#1f77b4', linewidth=2, linestyle='-')
plt.plot(x, y_cos, label='cos(x)', color='#ff7f0e', linewidth=2, linestyle='--')


# Add scatter plot with gradient colors
scatter = plt.scatter(x_scatter, y_scatter, c=y_scatter, cmap='viridis', alpha=0.7, edgecolors='w', s=100)

# Add colorbar for the scatter plot
plt.colorbar(scatter, label='Scatter Color Intensity')

# Add title and labels with custom fonts and sizes
plt.title('Beautiful Colorful Plot with Multiple Functions', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('x-axis', fontsize=14, fontweight='bold', color='#555555')
plt.ylabel('y-axis', fontsize=14, fontweight='bold', color='#555555')

# Add legend with custom location and frame styling
plt.legend(loc='upper right', frameon=True, framealpha=0.8, shadow=True, borderpad=1)

# Show grid with custom styling
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

# Customize the tick marks
plt.xticks(fontsize=12, fontweight='bold', color='#444444')
plt.yticks(fontsize=12, fontweight='bold', color='#444444')

# Enhance the plot's overall aesthetics
plt.tight_layout()
plt.show()

````

# Databases:  Google Trending Topics

````python
# Import necessary libraries
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

# Authenticate and create a BigQuery client
auth.authenticate_user() # Authenticate user for Google Colab

project_id = "secdata-379311"
client = bigquery.Client(project=project_id)



# Define the SQL query
query = """
SELECT
   refresh_date AS Day,
   term AS Top_Term,
   rank
FROM `bigquery-public-data.google_trends.top_terms`
WHERE
   rank = 1
   AND refresh_date >= DATE_SUB(CURRENT_DATE(), INTERVAL 2 WEEK)
GROUP BY Day, Top_Term, rank
ORDER BY Day DESC
"""

# Execute the query
query_job = client.query(query)

# Convert the results to a pandas DataFrame
df = query_job.to_dataframe()

# Display the DataFrame
df.head(15)  # Display the first few rows of the DataFrame
````
