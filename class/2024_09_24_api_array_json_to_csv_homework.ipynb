```python
import requests

url="https://earthquake.usgs.gov/fdsnws/event/1/query"


params = {
  "format" : "geojson",
  "starttime": "2024-09-10",
  "endtime" : "2024-09-20",
  "minmagnitude" : 4

}

resp=requests.get(url, params=params)

data=resp.json()['features']

data

```

```python
f={}

columnHeaders=[]
weather=[]

getHeader = True

for l in data:

  w = dict()

  for k,v in l.items():


    if isinstance(v,dict):
      for k1,v1 in v.items():
        w[k1]=v1
    elif isinstance(v,list):
      for k2,v2 in v[0].items():
        w[k2]=v2
    else:
      w[k]=v

  print(w)
  weather.append(w)


  if getHeader:
    columnHeaders=w.keys()
    getHeader=False
```
