```python
import requests

url="https://earthquake.usgs.gov/fdsnws/event/1/query"


params = {
  "format" : "geojson",
  "starttime": "2024-09-10",
  "endtime" : "2024-09-20",
  "minmagnitude" : 4

}

resp=requests.get(url, params=params)

data=resp.json()

data

```
