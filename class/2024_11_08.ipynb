```python
a = int(input('The first number= '))
b = int(input('The second number= '))

while True:
    c = a - b
    if c == 0:
        break
    elif b > c:
        a = b
        b = c
    else:
        a = c

print(f'The greatest common divisor is {a}')
```

```python
a = 17
b = 7*3

print("a={} b={} c={}".format(a,b,a-b))

while True:
    c = a - b
    if c == 0 or c==1:
        break
    elif b > c:
        a = b
        b = c
    else:
        a = c
    print("a={} b={} c={}".format(a,b,c))

print(f'The greatest common divisor is {a}')
```
