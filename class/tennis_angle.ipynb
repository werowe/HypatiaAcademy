```python

import math


def tallPlayerAdvantage(height):

  sl=6.4
  bl=11.8
  net=0.91

# arctangent is opposite/adjacent

  θsl=math.atan(net/sl)

  print("\nangle between service line and net")
  print("\nθ radians", θsl, "θ degrees ",math.degrees(θsl) )


# flatLine is the line such that you have line of site to the service line from the base line


# this is:
# opposite=math.tan(angle_radians) * adjacent

  flatLine = θsl * (sl + bl)

  print("\nflatLine", flatLine )

  # suppose a player is 1.83 meters.  if he fires a shot at the exact height of the net how far below the service line will it hit?

  # we need to know the adjacent side.  we only know the opposite side.  we do not know the hypoteneus length.  so we must find the tangent then we can calculate the length of the adjacent line and subtract the length of the service line to show the length of the green area.

  # know know that all the angles of a triangle add up to 180 dedgree.  so let's calculat the angle betwen this player's line and the line at the heigthe of the net

  #player = 1.83 # 6 feet
  #player = 1.96 # 6' 5"

  player=height

  aboveNet = player - net
  aboveNet

  tan = aboveNet / bl
  tan

  playerAngle=math.atan(tan)

  print("\nangle between player and top of net is ", playerAngle , "radians which is " , math.degrees(playerAngle) ,  " degrees")

  # angle between service line and flatline is

  θFlSl= math.pi / 2 - θsl

  math.degrees(θFlSl)

  # check.  since the angles in a triange sum to 180 it would also be

  θNetFlatLine=180 - 90 - math.degrees(θsl)
  θNetFlatLine

  # angle between brown player line and black flat line is:

  θPlayerFlat= 90 - θNetFlatLine - math.degrees(playerAngle)
  θPlayerFlat

  # so the angle between the player line and the net on the opposite side of the net from the player is

  θPlayerCourt = 180 - (θPlayerFlat + math.degrees(θFlSl)) - 90
  θPlayerCourt

  # I know the angle in degrees between the hypoteneuse and the adjacent line and the height of the opposite line how can I calculat the length of the adjacent line



  # Given angle in degrees and the height of the opposite side
  theta_degrees = 30  # Example angle between hypotenuse and adjacent side in degrees
  h_opposite = net      # Height of the opposite side

  # Convert the angle to radians for trigonometric calculations
  theta_radians = math.radians(θPlayerCourt)

  # Calculate the hypotenuse using the sine function
  hypotenuse = h_opposite / math.sin(theta_radians)

  # Calculate the adjacent side using the cosine function
  adjacent = hypotenuse * math.cos(theta_radians)

  # Print the result
  print(f"\nLength of the adjacent side: {adjacent}")

  # now subtract the service line to get the length of the green line.  this is the short player disadjantage meaning beyond line of site

  shortPlayerDisadvantage = adjacent - sl

  print("\nshort player disadntage", shortPlayerDisadvantage)

  return round(shortPlayerDisadvantage,2)





```

```python
playerHeightCm=1.83

print("\n\nshort player disadvantage is ",tallPlayerAdvantage(playerHeightCm), "for a player of height", playerHeightCm)
```

Given:

- The angle $ \theta $ between the hypotenuse and adjacent side (in degrees)
- The height of the opposite side $ h_{\text{opposite}} $

### Step 1: Calculate the hypotenuse using the sine function

$$
\text{hypotenuse} = \frac{h_{\text{opposite}}}{\sin(\theta)}
$$

### Step 2: Calculate the adjacent side using the cosine function

$$
\text{adjacent} = \text{hypotenuse} \times \cos(\theta)
$$

### Combined Formula:

Substituting for the hypotenuse:

$$
\text{adjacent} = \frac{h_{\text{opposite}}}{\sin(\theta)} \times \cos(\theta)
$$


```python
import numpy as np

heights = np.arange(2.57, 1.57 - 0.1, -0.10)

disadvantage = np.array([tallPlayerAdvantage(x) for x in heights])
```

```python
np.column_stack([heights,disadvantage])
```

```python
import matplotlib.pyplot as plt


# Create the plot
plt.plot(disadvantage, heights, marker='o')

# Add labels
plt.xlabel('Disadvantage')
plt.ylabel('Height')

# Add a title
plt.title('Height vs Disadvantage')

# Label each point with its (x, y) coordinates, rounding the x-value to 2 decimal places
for x, y in zip(disadvantage, heights):
    plt.text(x + 0.25, y, f'({round(x, 2)}, {round(y, 2)})', fontsize=9, ha='left',color='purple')  # Shift text slightly to the right

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Fit a polynomial of degree 2 (quadratic curve)
coefficients = np.polyfit(heights, disadvantage, 2)

# Generate the polynomial equation
poly_eq = np.poly1d(coefficients)

# Print the polynomial equation
print(f"The polynomial equation is: \n{poly_eq}")

# Plot the original data
plt.scatter(heights, disadvantage, color='red', label='Data points')

# Plot the polynomial curve
x_curve = np.linspace(min(heights), max(heights), 500)
y_curve = poly_eq(x_curve)
plt.plot(x_curve, y_curve, label='Fitted curve')

# Labels and title
plt.xlabel('height')
plt.ylabel('disadvantage')
plt.title('Curvilinear Relationship')

plt.legend()
plt.show()
```
