```python
s=[10,9,8,7]

#Pavlo


print("Pavlo start", s, "\n")

print("\n checking 0 and 1", s)
if s[1] < s[0]:
  s[1] , s[0] = s[0] , s[1]


print("\n checking 0 and 2", s)
if s[2] < s[0]:
  s[2] , s[0] = s[0] , s[2]

print("\n checking 0 and 3", s)
if s[3] < s[0]:
  s[3] , s[0] = s[0] , s[3]


print("\n checking 1 and 2", s)
if s[2] < s[1]:
  s[2] , s[1] = s[1] , s[2]

print("\n checking 1 and 3", s)
if s[3] < s[1]:
  s[3] , s[1] = s[1] , s[3]


print("\n checking 2 and 3", s)
if s[3] < s[2]:
  s[3] , s[2] = s[2] , s[3]


print("\nfinish", s)
```

```python
# remove hard-coding.  write algorithm to work with any length list
# this is the classic bubble-sort



import random

s=[10,9,8,7]

#works with words too
#s=["Pavlo", "Oleski", "Julia", "Alex"]



#s=[]
#for i in range(5):
#  s.append(random.randint(1,10))

print("\nstart sorting", s)

loops = int(len(s) - 1)

for j in range(loops):
  print("\n===============start loop========", j)

  for i in range(loops):
    print("\nchecking {0} < {1}\n".format(i, i+1))
    if s[i] > s[i+1]:
      s[i], s[i+1] = s[i+1], s[i]

  print("\n============end loop============", s)

print("\ndone sorting", s)


```
