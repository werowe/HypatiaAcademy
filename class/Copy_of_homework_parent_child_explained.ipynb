```python
# you need some common element to link two tables

%%sql

CREATE TABLE parent
  (
    parent_id INTEGER PRIMARY KEY AUTOINCREMENT,
    parent_name text,
    date_created text
  );

CREATE TABLE child
  (
    child_id INTEGER PRIMARY KEY AUTOINCREMENT,
    parent_id INTEGER,
    child_name text,
    date_created text
  );
```

```python
%%sql


CREATE TABLE orders
  (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    product_name text,
    sales_amount real
  );

```

```python
%%sql
INSERT INTO orders(product_id, product_name, sales_amount) values("456", "Snickers", 15);
INSERT INTO orders(product_id, product_name, sales_amount) values("456", "Snickers", 25);
```

```python
%%sql

select sum(sales_amount), product_name from orders
group by product_id
```

```python
%%sql

select count(sales_amount), product_name from orders
group by product_id
```

```python
%%sql

select ave(sales_amount), product_name from orders
group by product_id
```
