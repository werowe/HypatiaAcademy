```python
# Example 1 of reshape: Reshaping arr_random (4x5) to (2,10)
r_ex_1 = arr_random.reshape(2, 10)
print("Reshaped Example 1 (2x10 array):")
print(r_ex_1)
print(r_ex_1.shape)

# Example 2 of reshape: Reshape arr_random (4x5) to (1, -1) (flatten to a single row)
r_ex_2 = arr_random.reshape(1, -1)
print("\nReshaped Example 2 (flattened to 1 row):")
print(r_ex_2)


print(r_ex_2.shape)
```

```python
import numpy as np

def label(x):
    return "even" if x % 2 == 0 else "odd"

vlabel = np.vectorize(label)
a = np.array([1, 2, 3, 4])
res = vlabel(a)   # array(['odd', 'even', 'odd', 'even'], dtype='<U4')
```
