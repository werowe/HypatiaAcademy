```python

from enum import Enum

class Color(Enum):
  BLUE = 1,
  BLACK = 2,
  RED = 3,
  YELLOW = 4,
  BROWN = 5,
  GREEN = 6
```

```python
# base class

class Animal:

   # constructor.  that means the method used to create the object

   def __init__(self, weight,  age,  hasFur=False, isDangerous=False,name=None, color=None):

      # self means inside the class.  weight is a parameter
      self.weight = weight
      self.name = name
      self.color = color


   def getVaccinated(self):

      self.isVacinated = True



# derived class

class Dog(Animal):


       hasEyes = True

       def __init__(self, weight,  age, breed, license, name=None, doesBite=False):

           super().__init__(weight, age)
           self.breed = breed
           self.hasFur = True
           self.doesBite = False
           self.name = name
           self.license = license

          # sound in decimals
           self.volume = 60

           # represent
       def __str__(self):

           return f"This dogs is named {self.name} and bites {self.doesBite}"

       def __eq__(self, license):

          return self.license == license

class Cat(Animal):

     def __init__(self, weight,  age):

          self.hasFur = True
          self.volume = 3

class Turtle(Animal):

       def __init__(self, weight,  age, type, color=None, name=None, doesBite=False):

           super().__init__(weight, age)
           self.type = type
           self.hasFur = True
           self.doesBite = True
           self.name = name


          # sound in decimals
           self.volume = 60

           # represent
       def __str__(self):

           return f"This dogs is named {self.name} and bites {self.doesBite}"



       def __eq__(self, license):

          return self.license == license
```
