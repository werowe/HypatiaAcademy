````python
#Sunrise, sunset at the specific coordinates (I put Kyiv here)


url = 'https://api.sunrise-sunset.org/json?lat=50.450001&lng=30.523333'

# Look in the documentation to see if you can get multiple days on sunrise information in one API call, meaning can date be a range?

params = {'lat': 50.45,
               'lon': 30.52,
          'date': '1997-09-17'
               }


response = requests.get(url, params=params)

print(json.dumps(response.json(), indent=4))


# we want just the key['result'] as the important data is below that.

sunrise=response.json()['results']

#There is no info about how get the multiple days at once
#We can do it only doing seperate API requests
````

````python
# we need the data to write to be a list.  because here we only have one row which is the sunrise for one day.  But normall you would have many rows.

days=[]

# add the sunsrise dictionary to the list

days.append(sunrise)
````

````python
import csv


# create column headers

columnHeaders=sunrise.keys()


# write the data to a text csv (comma-delimited file) using the csv.DictWriter API

with open('sunrise.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=columnHeaders)
    writer.writeheader()  # Write header
    writer.writerows(days)  # Write data rows
````
