# Symbols to Represent Words, no Words, Special Characters, Position, etc.


* .	Any character except newline
* ^	Start of string / line
* $	End of string / line
* \*	0 or more repetitions
* \+	1 or more repetitions
* ?	0 or 1 repetition (optional)
* []	Match any character in brackets
* Match any character NOT in brackets
* \d	Digit (0–9)
* \D	Not a digit
* \w	Word character (alphanumeric or '_')
* \W	Not a word character
* \s	Whitespace (space, tab, etc.)
* \S	Not a whitespace character
* \b	Word boundary
* \B	Not a word boundary
* ( )	Capture group
* \	Escape special character
* {n}	Exactly n repetitions
* {n,}	n or more repetitions
* {n,m}	Between n and m repetitions

# Counts/Logic Symbols


| Symbol  | Meaning                        | Example      |
|---------|--------------------------------|--------------|
| ?       | 0 or 1 (optional)              | colou?r      |
| *       | 0 or more (any count)          | ab*c         |
| +       | 1 or more                      | ab+c         |
| .       | any single character           | a.b          |
| {n}     | exactly n times                | a{3}         |
| {n,}    | n or more times                | a{2,}        |
| {n,m}   | between n and m times          | a{2,5}       |


# regular expression

This is a way to match text



```python
import re

dt = "2024-10-12"


# python like java requires \\ two of those
# instead of 1

pattern="\\d{4}-\\d\\d-\\d{2}"

if re.match(pattern, dt) is None:
  print(" go back and put valid date in YYYY-MM-DD format")
else:
  print("looks good")


```

```python
import re

word = "Yeremii"


# python like java requires \\ two of those
# instead of 1

pattern="Yeremi{2}"

if re.match(pattern, word) is None:
  print(f" {word} did not match  {pattern}")
else:
  print("looks good")


```

```python
import re

word = "YEREMII"


# python like java requires \\ two of those
# instead of 1

pattern="[0-9][a-zA-Z]*"

if re.match(pattern, word) is None:
  print(f" {word} did not match  {pattern}")
else:
  print("looks good")


```

```python
import re

word = "YEREMII"


#  ? means optional
# {n} means exactly n times
# means * any number times


pattern="[0-9]?[a-zA-Z]*"

if re.match(pattern, word) is None:
  print(f" {word} did not match  {pattern}")
else:
  print("looks good")
```

## Ukrainian Phone Numbers

Example Ukrainian phone number:
+380 50 123 45 67

What this demonstrates:

* Country code: +380

* Network/operator code: 50 (two digits)

* Local subscriber number: 7 digits grouped as 123 45 67 or 1234567 depending on formatting

* Optional separators: spaces (also commonly used are spaces, dashes, or nothing)

Common formats you’ll see:

* +380 50 123 45 67

* +380(50)123-45-67

* +380501234567

* 050 123 45 67

* (050) 123-45-67

Notes on pattern design:

* Start with an optional plus sign and country code +380

* Follow with a two- to four-digit area/operator code

* Then a seven-digit local number, which can be grouped in various ways

* Allow spaces, dashes, or no separators

* Be mindful of optional parentheses around the area/operator code
