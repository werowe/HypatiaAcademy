```python
import json
import requests


api_key="563f19204c5b4751e6f9eb0b00aaf357"
URL="https://api.openweathermap.org/data/2.5/forecast"

params={"lat":"49.4413",
        "lon":"32.0643",
        "exclude":"current,minutely,hourly,alerts",
        "appid": api_key,
        "units":"metric",
        "cnt":"56",
        "lang":"ua"}

response=requests.get(URL,params=params)
result=response.json()
print(json.dumps(result, indent=4, sort_keys=True))

```

```python
f={}

columnHeaders=[]
weather=[]

getHeader = True

for l in result['list']:

  w = dict()

  for k,v in l.items():


    if isinstance(v,dict):
      for k1,v1 in v.items():
        w[k1]=v1
    elif isinstance(v,list):
      for k2,v2 in v[0].items():
        w[k2]=v2
    else:
      w[k]=v

  print(w)
  weather.append(w)


  if getHeader:
    columnHeaders=w.keys()
    getHeader=False

```

```python
columnHeaders
```

```python

import csv

with open('forecast.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=columnHeaders)
    writer.writeheader()  # Write header
    writer.writerows(weather)  # Write data rows
```
