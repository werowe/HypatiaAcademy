```python


import json

import requests


url="https://api.openweathermap.org/data/2.5/forecast"


params = {'lat': 50.45,
               'lon': 30.52,
               'appid' : 'bd5e378503939ddaee76f12ad7a97608',
               'units' : 'metric',
               'exclude' : ['daily', 'hourly']
               }


response = requests.get(url, params=params)

print(json.dumps(response.json(), indent=4))
```

```python
forecast=response.json()['list']
```

```python
for f in forecast:
  print(f['dt_txt'])
```

```python
w=[]

wHeaders=['dateTime', 'temp', 'temp_min', 'temp_max', 'weather', 'windSpeed', 'windDirection']
for f in forecast:

  d={}

  d['dateTime']= f['dt_txt']
  d['temp']=f['main']['temp']
  d['temp_min']=f['main']['temp_min']
  d['temp_max']=f['main']['temp_max']
  d['weather']=f['weather'][0]['main']
  d['windSpeed']=f['wind']['speed']
  d['windDirection']=f['wind']['deg']

  w.append(d)

```

```python

import csv



with open('forecast.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=wHeaders)
    writer.writeheader()  # Write header
    writer.writerows(w)  # Write data rows
```
