```python
from datetime import datetime

from google.colab import drive
import shutil

drive.mount('/content/drive')


from sqlalchemy import (
    Column,
    Integer,
    String,
    BigInteger,
    create_engine,
    text,
)
from sqlalchemy.orm import declarative_base, sessionmaker

# Path to your SQLite database file (same as db_path)
DB_URL = "sqlite:////content/drive/MyDrive/inventory.db"

engine = create_engine(DB_URL, echo=False, future=True)

# Enable foreign keys for SQLite
with engine.connect() as conn:
    conn.execute(text("PRAGMA foreign_keys = ON;"))

Base = declarative_base()


class Customer(Base):
    __tablename__ = "customers"

    customernumber = Column(Integer, primary_key=True, autoincrement=True)
    customername = Column(String)
    phonenumber = Column(String)
    postalcode = Column(String)
    datecreated = Column(BigInteger)  # store Unix timestamp like your strftime('%s','now')
    email = Column(String)

    def __repr__(self):
        return (
            f"Customer(customernumber={self.customernumber}, "
            f"customername={self.customername!r}, email={self.email!r})"
        )


# Create tables (equivalent to your CREATE TABLE statement)
Base.metadata.create_all(engine)

SessionLocal = sessionmaker(bind=engine, autoflush=False, autocommit=False, future=True)


def main():
    session = SessionLocal()

    try:
        # INSERT: add a customer (John Doe)
        john = Customer(
            customername="John Doe",
            phonenumber="555-1234",
            postalcode="12345",
            datecreated=int(datetime.utcnow().timestamp()),
            email="johndoe@example.com",
        )
        session.add(john)

        # INSERT: add another customer (Fred Smith)
        fred = Customer(
            customername="Fred Smith",
            phonenumber="555-1234",
            postalcode="12345",
            datecreated=int(datetime.utcnow().timestamp()),
            email="fred@example.com",
        )
        session.add(fred)

        session.commit()

        # SELECT * FROM customers
        customers = session.query(Customer).all()
        print("Customer Records:")
        for c in customers:
            print(
                c.customernumber,
                c.customername,
                c.phonenumber,
                c.postalcode,
                c.datecreated,
                c.email,
            )

    except Exception as e:
        session.rollback()
        print("Error:", e)
    finally:
        session.close()


if __name__ == "__main__":
    main()
```
