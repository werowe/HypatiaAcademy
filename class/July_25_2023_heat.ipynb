````python
import pandas as pd

df=pd.read_json('paphosWeather2023.json')

df
````

````python
df.columns


'''
'''
"main": {
		"temp": 22.99,
		"temp_min": 20.6,
		"temp_max": 24.23,
		"feels_like": 23.51,
		"pressure": 1009,
		"humidity": 83,
		"dew_point": 19.95
	}
'''
````

````python
heat=df[['dt_iso','main']]
heat['main']


````

````python
# map means run some functions over every row
# lamdba means create an in-line function


def getTemperature(d):
    return d["temp"]

heat['temp']=df['main'].map(getTemperature)

heat.set_index("dt_iso",inplace=True)
````
