```python
import re

e=[]



pattern = r"\b(Chest|Review)\b"


totalLessions = 0

for i in pathSectioned['pathSectioned']['currentCourse']:
 cnt = 1

 for j in (i['units']):
    lessons = []
    d = {}
    section=i['index']+1
    d['section']=section
    if section < 4:
      if d['section'] != prevSection:
        cnt = 1
      d['unit']= cnt
      d['teachingObjective'] = j['teachingObjective']

      lessons = []
      for l in j['levels']:
        skip = re.findall(pattern, l["debugName"])
        if len(skip) == 0:
          lessons.append(l["totalSessions"])
          totalLessions = totalLessions + l["totalSessions"]
      d['lessons'] = lessons
      e.append(d)
      cnt = cnt + 1
      prevSection=d['section']

print("totalLessions=", totalLessions)

df=pd.DataFrame(e)
pd.set_option('display.max_rows', None)


exploded=df.explode("lessons")

```
