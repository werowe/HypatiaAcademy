```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Years and PC sales data
years = np.array([1982, 1985, 1990, 1995, 2000, 2001, 2007, 2009, 2012, 2015])
sales = np.array([2.5, 15, 70, 270, 850, 1350, 1600, 1750, 1900, 2100])  # PC sales in millions

# Price data for DOS/Windows in USD
prices = np.array([40, 99, 149, 209.95, 319, 499, 399, 319, 199, 119])

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot PC sales on the primary y-axis
ax1.plot(years, sales, marker='o', label='PC Sales', color='b')
ax1.set_xlabel('Year')
ax1.set_ylabel('PC Sales (Millions)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add a secondary y-axis for prices
ax2 = ax1.twinx()
line_prices, = ax2.plot(years, prices, marker='s', label='DOS/Windows Price', color='magenta')
ax2.set_ylabel('Price of DOS/Windows (USD)', color='magenta')
ax2.tick_params(axis='y', labelcolor='magenta')

# Add vertical lines and annotations for key events
plt.axvline(x=2001, color='r', linestyle='--')
plt.text(2001 + 0.2, ax1.get_ylim()[1] * 0.05, '2001 US Antitrust Case', color='r', rotation=90,
         verticalalignment='bottom')

plt.axvline(x=2006, color='purple', linestyle='--')
plt.text(2006 + 0.2, ax1.get_ylim()[1] * 0.05, 'Amazon EC2 Launch (2006)', color='purple', rotation=90,
         verticalalignment='bottom')

plt.axvline(x=2007, color='g', linestyle='--')
plt.text(2007 + 0.2, ax1.get_ylim()[1] * 0.05, 'First iPhone (2007)', color='g', rotation=90,
         verticalalignment='bottom')

plt.axvline(x=2008, color='b', linestyle='--')
plt.text(2008 + 0.2, ax1.get_ylim()[1] * 0.05, 'Android Release (2008)', color='b', rotation=90,
         verticalalignment='bottom')

# Add titles
plt.title('PC Sales and DOS/Windows Prices Over Time')
plt.grid(True)

# Initialize markers for animation
marker_sales_line = ax1.plot([], [], 'bo', markersize=10)[0]   # Blue marker for PC sales
marker_prices_line = ax2.plot([], [], 'mo', markersize=10)[0] # Magenta marker for prices

# Animation function
def update(frame):
    # Update marker positions based on the current frame
    marker_sales_line.set_data([years[frame]], [sales[frame]])  # Wrap scalars in lists
    marker_prices_line.set_data([years[frame]], [prices[frame]])  # Wrap scalars in lists
    return marker_sales_line, marker_prices_line

# Create the animation
frames = len(years)
ani = FuncAnimation(fig, update, frames=frames,
                    interval=500) # Interval in milliseconds

# Save the animation as a GIF file
writer = PillowWriter(fps=2) # Frames per second
ani.save("animated_markers.gif", writer=writer)

# Show the plot (optional)
plt.show()

```
