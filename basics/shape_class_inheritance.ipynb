```python
import math

class Shape:


    def __init__(self, points):

      self.points=points

    # this based class function can be used on any derived class:

    def returnXYPoints(self):

        x = [point[0] for point in self.points]
        y = [point[1] for point in self.points]

        return x,y

    #show the advantage of polymorphism

    def rotate(self, degrees):

      # Convert degrees to radians
      radians = math.radians(degrees)

      # Rotation matrix components
      cos_theta = math.cos(radians)
      sin_theta = math.sin(radians)

      # Rotate each point in the list
      rotated_points = []
      for (x, y) in self.points:
        x_new = x * cos_theta - y * sin_theta
        y_new = x * sin_theta + y * cos_theta
        rotated_points.append((x_new, y_new))

      return rotated_points
```

The mathematical formula for rotating a point $(x, y)$ around the origin by an angle $\theta$ (in degrees) is:

$$
x' = x \cdot \cos(\theta) - y \cdot \sin(\theta)
$$

$$
y' = x \cdot \sin(\theta) + y \cdot \cos(\theta)
$$

Where $(x', y')$ are the new coordinates of the rotated point, and $\theta$ is the angle of rotation.
