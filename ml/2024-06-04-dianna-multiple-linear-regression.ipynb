```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from pylab import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

Marvel = pd.read_csv("https://raw.githubusercontent.com/dievikson/HypatiaAcademy/master/ml/Marvel%20Movies.csv")
```

```python
movies = Marvel[['critics % score', '% budget recovered','1st vs 2nd weekend drop off']].copy()
```

```python
y = Marvel.iloc[:,-1].values
```

```python
import re

def toNumber(n):

    s=re.sub(r'[^\d\.]+', '', n)
    i=int(s)
    return i/100
```

```python
toNumber("-45%")
```

```python
movies
```

```python
movies['x2']=movies['x2'].apply(toNumber)
movies['x1']=movies['x1'].apply(toNumber)
movies['y']=movies['y'].apply(toNumber)
```

```python
movies
```

```python
m=movies.to_numpy()
m
```

```python
y=m[:,0]

y
```

```python

x1x2=m[:,1:3]
x1x2
```

```python
# when you want do to a prediction and you have 1,000 data samples.  the normal procedure
# is to take maybe 40% to use to train the model.  train means find the coefficients.
# that means find m1 m2 ...

# and then you use the remaining part to test it.  test it means to see how accurate it is.




#xtrain,xtest=np.array_split(x, 2)
#ytrain,ytest=np.array_split(y, 2)


reg = LinearRegression()
reg.fit(x1x2, y)

# linear regression solves this for m1 and m2

# y = m1 * x1 + m2 * x2 + b

# this is our predictive model
```

```python
np.set_printoptions(suppress=True, precision=10)

print('Coefficients: \n', reg.coef_)
print('\n Intercept: \n', reg.intercept_)
```

```python
# means we IF this model is correct then we estimate the 'critics % score',  on:

# '% budget recovered','1st vs 2nd weekend drop off

# 'critics % score' = y
# '% budget recovered' = x1
# '1st vs 2nd weekend drop off'

# y = 0.0222460491 * x1 -0.9080425153 * x2 +  1.2385964518236987
```

```python
lastRow = [ 0.84,  7.14,  0.65]

# these are the coefficients that the linear regression algorithm found:
# reg = LinearRegression()<----create the model or the algorithm
# reg.fit(x1x2, y)<----fit a line, or find the that goes down the middle
# using the least square errors technique


m1 = 0.0222460491
m2 = -0.9080425153
b = 1.2385964518236987


x1 = 7.14
x2 = 0.65

ypred= m1 * x1 + m2 * x2 + b
ypred
```

```python
# Make predictions
y_pred = reg.predict(x1x2)

# Calculate R-squared
r2 = r2_score(y, y_pred)
print(f"R-squared: {r2}")

```

```python
# b is y intercept when we have one variable x and we are solving simple linear regression

# which is y = mx + b

# b is the point on the graph (b,0)
# this is the y intercept
```

```python
import statsmodels.api as sm

# Add a constant to the model (intercept)
X_train_sm = sm.add_constant(x1x2)

# Fit the model using statsmodels
model_sm = sm.OLS(y, X_train_sm).fit()

# Print the summary to get p-values, R-squared, and other statistics
print(model_sm.summary())
```
