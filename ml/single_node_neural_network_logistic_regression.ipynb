## Forward Pass

- **Input:** $x$
- **Parameters:** $w, b$
- **Linear Combination:**
  $$
  z = w^T x + b
  $$
- **Activation:**
  $$
  a = \sigma(z)
  $$
  where $\sigma(z)$ is the sigmoid function.
- **Loss Function:**
  $$
  \mathscr{L}(a, y)
  $$






## Loss Function

The **output of the networks final neuron is a real number between 0 and 1 (lets call it `a`), and the true target value `y` is an integer, either 0 or 1**. This might look inconsistent at first glance, but it's by design and is fundamental to how probabilistic classification works.

**Heres why this is not a problem and, in fact, necessary:**

- **The output `a` is interpreted as the predicted _probability_ that the input belongs to class 1**for example, if `a = 0.9`, the model estimates there is a 90% chance the input is in class 1.
- **The observed value `y` is the true label**, which is always exactly 0 or 1.

### Why compare a probability with a binary value?


$$
da= \frac{\partial}{\partial a} \mathscr{L}(a, y) = -ylog(a) -(1-y)log(1-a)
$$

$$
=-\frac{y}{a} + \frac{1 - y}{1 - a}
$$

---

- If `y = 1`, the loss simplifies to `-log(a)`.
    - The closer `a` is to 1, the smaller this loss**rewarding high probability for the correct class**.
- If `y = 0`, the loss simplifies to `-log(1 - a)`.
    - The closer `a` is to 0, the smaller this lossagain, **rewarding correct prediction**.

**This loss encourages the network to output a probability close to the actual observed class**. The fact that you're "comparing" a real-valued probability `a` with a binary `y` is not a mismatch: the cross-entropy formula naturally expresses the likelihood of seeing the actual label `y` given the model's prediction `a`.

> This is functionally equivalent to maximizing the likelihood that the model assigns to the observed data: we want the model to be as "confident" (probability close to 1 for `y=1`, close to 0 for `y=0`) and as accurate as possible for each example.

## Gradient Descent

Repeat:

1. **Compute predictions** $(\hat{y}^{(i)},\ i=1, \ldots, \ m)$
2. **Compute gradients:**
    - $dW^{[1]} = \frac{\partial J}{\partial W^{[1]}}$
    - $db^{[1]} = \frac{\partial J}{\partial b^{[1]}}$
3. **Update parameters:**
    - $W^{[1]} := W^{[1]} - \alpha \ dW^{[1]}$
    - $b^{[1]} := b^{[1]} - \alpha \ db^{[1]}$

Repeat for all layers $l$.
