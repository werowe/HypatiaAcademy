```python
import pandas as pd

df = pd.read_csv("sample_data/california_housing_test.csv")
```

```python
df
```

```python
df.columns
```

```python
# center
# 'longitude', 'latitude'

# is_in_center
```

```python
Y = df['median_house_value'].values
Y
```

```python
df.drop('median_house_value', axis=1, inplace=True)
```

```python
X=df.values
X
```

```python
X.shape
```

```python
Y=Y.reshape(-1,1)
```

```python
Y.shape
```

```python
import numpy as np
import sklearn




linear_model = sklearn.linear_model

lr = linear_model.LinearRegression()

lr.fit(X, Y)


```

```python
m = lr.coef_      # slope(s)
b = lr.intercept_ # intercept

np.set_printoptions(suppress=False)

print("m coefficient:", m)
print("b bias:", b)
```
