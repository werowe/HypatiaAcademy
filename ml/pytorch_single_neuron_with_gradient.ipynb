# Training a neural network
Means to update weights and bias to get the smallest loss.


Just to try to show you how it works we were calculation gradients manually.

To do that you need to understand calculus.

Use the gradient ability of my torch to automatically do it.

If the gradient of dl/dw > 0 then you need to reduce the weight


1.  forward pass
    - z = wx + b
    - sigmoid (z)  (makes prediction, between 0 and 1)
    - loss function (check the error.  y_hat - y
    - calculate the derivative or the gradient of the loss function with respect to w and b
2.  back propagation
    - w = w - lr * dl/dw  now update the weights based up the dl/dw
    - b = b - lh * dl/db  updat the bia based on dl/db
    - lr is learning rate 0.01


```python
import torch

# -----------------------------
# Input data: one training example
# -----------------------------
x = torch.tensor([1.0, 2.0, 3.0])   # features
y = torch.tensor(1.0)                # true label: 0 or 1

# -----------------------------
# Parameters of one neuron
# -----------------------------

# start off just make up some values for w and b
# guess
# goal is to reduce the loss until it gets to be small
w = torch.tensor([0.1, 0.2, 0.3], requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

print("\n inital weights", w)
print("\n inital bias", b)

# x are our features
# like blood sugar, blood pressure
print("\n x will never change", x)


# -----------------------------
# Forward pass
# -----------------------------
z = torch.dot(w, x) + b              # linear part
y_hat = torch.sigmoid(z)             # activation function

# Binary cross-entropy loss
loss = -(y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat))

print("\ndot product z =", z.item())
print("\n prediction y_hat =", y_hat.item())
print("\nloss =", loss.item())

# -----------------------------
# Backward pass
# -----------------------------

# when you run this it automatically caculates the derivative
#w.grad, b.grad

loss.backward()

print("\nderivative of loss with respect to weight dw =", w.grad)
print("\n derivate of loss with respect to bias db =", b.grad)

# -----------------------------
# Update parameters manually
# -----------------------------
learning_rate = 0.1

with torch.no_grad():
    w -= learning_rate * w.grad
    b -= learning_rate * b.grad

    # Clear gradients before next training step
    w.grad.zero_()
    b.grad.zero_()

print("\n updated weight w =", w)
print("\n updated bias b =", b)
```

```python
learning_rate = 0.1
# -----------------------------
# Input data: one training example
# -----------------------------
x = torch.tensor([1.0, 2.0, 3.0])   # features (e.g., input values for a neuron)
y = torch.tensor(1.0)                # true label (the desired output, 0 or 1 for binary classification)

# -----------------------------
# Parameters of one neuron
# -----------------------------

# Initial random weights for the neuron, requires_grad=True tracks gradients for these tensors
w = torch.tensor([0.1, 0.2, 0.3], requires_grad=True)
# Initial random bias for the neuron, requires_grad=True tracks gradients
b = torch.tensor(0.0, requires_grad=True)

print("\n inital weights", w)
print("\n inital bias", b)

# The input features 'x' remain constant during training
print("\n x will never change", x)



epochs = 100 # Number of times to run the training process

for i in range(epochs):

# -----------------------------
# Forward pass: Calculate the prediction
# -----------------------------
  # Linear combination of inputs and weights, plus bias
  z = torch.dot(w, x) + b
  # Apply sigmoid activation function to get a prediction between 0 and 1
  y_hat = torch.sigmoid(z)

# Binary cross-entropy loss: Measures the error between prediction (y_hat) and true label (y)
  loss = -(y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat))

# Print the loss for each epoch to see how it decreases over time
  print("\nloss =", loss.item())

# -----------------------------
# Backward pass (Backpropagation): Calculate gradients
# -----------------------------

  # Computes gradients of the loss with respect to all tensors that have requires_grad=True
  loss.backward()

# -----------------------------
# Update parameters: Adjust weights and bias to reduce loss
# -----------------------------

  # Temporarily disable gradient tracking for parameter updates
  with torch.no_grad():
    # Update weights: subtract learning_rate * gradient of weights
    w -= learning_rate * w.grad
    # Update bias: subtract learning_rate * gradient of bias
    b -= learning_rate * b.grad

    # Clear gradients after updating parameters to prepare for the next iteration
    w.grad.zero_()
    b.grad.zero_()
```
