linear regression means to solve y = mx + b

simplea linear regression means we have 1 independent variable x and 1 2 dependent variables

```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn as sklearn

x = np.array([0, 1, 2, 3, 4, 5, 6])

# observed values

y = np.array([0, 2.3, 4.2, 5.6, 6.7, 6.2, 9.12])
```

```python
# I want to train the model

from sklearn.linear_model import LinearRegression

reg = LinearRegression()

# train means solve for m and b the equation y = mx + b

x = x.reshape(-1, 1)
# because it 0 dimensions if you don't

# the input x must be of shape(number of inputs, number of features)


reg.fit(x, y)
```

```python
x[0:]
```

After training the model, you typically evaluate its performance using metrics like Mean Squared Error (MSE).

First, let's predict the `y` values using the trained model on the original `x` data.

```python
from sklearn.metrics import mean_squared_error

# Predict y values using the trained model
y_predicted = reg.predict(x)

print(f"Predicted y values: {y_predicted}")
```

```python
np.column_stack((x, y))
```

Now, we can calculate the Mean Squared Error between the actual `y` values and the `y_predicted` values.

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y, y_predicted)

print(f"Mean Squared Error: {mse}")
```

Regression finds the least square error.  

```python
print(f"y = {reg.coef_[0]}x + {reg.intercept_}")
```

```python
plt.scatter(x, y, label='Original Data')
plt.plot(x, y_predicted, color='red', label='Linear Regression Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression: Original Data vs. Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```
