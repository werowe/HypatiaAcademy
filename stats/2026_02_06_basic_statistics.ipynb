```python
import numpy as np
import matplotlib.pyplot as plt

# Example: turn your histogram counts into "share per bin" (pie slices)
cnts, bins = np.histogram(weights2, bins="auto")

labels = [f"[{bins[i]:.2g}, {bins[i+1]:.2g})" for i in range(len(cnts))]

# Optional: drop empty bins so you don't get zero-size slices
mask = cnts > 0
cnts2 = cnts[mask]
labels2 = [lab for lab, ok in zip(labels, mask) if ok]
labels = [f"[{bins[i]:.2f}, {bins[i+1]:.2f})" for i in range(len(cnts))]
# or 3 decimals:
# labels = [f"[{bins[i]:.3f}, {bins[i+1]:.3f})" for i in range(len(cnts))]


colors = plt.cm.viridis(np.linspace(0.15, 0.95, len(cnts2)))

fig, ax = plt.subplots(figsize=(7, 5))
ax.pie(
    cnts2,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"edgecolor": "white", "linewidth": 1.0},
)
ax.set_title("Share of weights by bin")
ax.axis("equal")  # keep it circular
fig.tight_layout()
plt.show()
```
