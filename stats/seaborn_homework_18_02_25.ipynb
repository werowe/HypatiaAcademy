```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# getting the dataset
penguinsData = sns.load_dataset('penguins')

# creating a custom color palette to use in the barplot
# seaborn will now use this one as the default color palette for every barplot
customPalette = sns.color_palette(['#2a57de', '#cb32e3'])
sns.set_palette(customPalette)

# creating the barplot
sns.barplot(
    x = 'species',
    y = 'body_mass_g', # if we set the y axis to the mass parameter, it will automatically calculate the means for each sex in each species
    hue = 'sex', # telling the seaborn to divide all species' data into male and female columns
    data = penguinsData, # setting the dataset
    order = ['Adelie', 'Chinstrap', 'Gentoo'], # setting the order of species on the x axis
    capsize = 0.1, # size of caps that represent possible data deviations
    saturation = 0.9, # opacity of colors
    errorbar='sd' # setting the caps to show one std as a possible data deviation
)

# showing the barplot
plt.show()
```
