```python
import pandas as pd
import numpy as np


def unwrap(y):

  totalSessions=[]

  for i in y['levels']:

     totalSessions.append(i['totalSessions'])

  d = {
      "totalSessions": totalSessions,
  }

  return totalSessions

cu="https://raw.githubusercontent.com/werowe/HypatiaAcademy/master/stats/course.json"

# get with wget first
#uu="https://raw.githubusercontent.com/werowe/HypatiaAcademy/master/stats/user.json"

course = pd.read_json(cu)


pathSectioned=pd.DataFrame(course.loc[['pathSectioned'],['currentCourse']])
pathSectioned=pathSectioned.T
units=pd.DataFrame(pathSectioned.loc[:,"pathSectioned"][0])
units=units.iloc[0:3]

a=units.loc[:,['index','debugName', 'completedUnits', 'totalUnits', 'units']]
a.rename(columns={'index': 'idx'}, inplace=True)
a

df = a

exploded_df = df.explode('units')

# Reset the index to maintain the original index as a column
exploded_df.reset_index(drop=True, inplace=True)

exploded_df['levels']=exploded_df['units'].apply(unwrap)

df = exploded_df.drop('units', axis=1)

exploded_df.reset_index(drop=True, inplace=True)

df=df.explode("levels")

print("total courses", df['levels'].sum())

```
