```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a barplot
sns.barplot(
    x="day",
    y="total_bill",
    hue="sex",
    data=tips,
    estimator=sum,
    errorbar="ci",   #confidence interval
    palette="pastel",
    order=["Thur", "Fri", "Sat", "Sun"],
    capsize=0.2,
)

# Add title and show plot
plt.title("Total Bill by Day and Gender")
plt.show()
```
