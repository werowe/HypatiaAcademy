```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [1, 3, 5, 7, 9]

# Create subplots with specified height ratios
fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})

# Plot data
ax1.plot(x, y)
ax2.plot(y, z)

# Set titles and labels
ax1.set_title('Top Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

ax2.set_title('Bottom Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.5)  # Increase vertical space between subplots

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


arr=np.array([1,1,2,2,2,2,2,3,3,3,3,3,4,4,4,4])

# Create a histogram with bins for integers
bins = np.arange(arr.min(), arr.max() + 2)  # Add 2 to include the last integer bin

counts, bin_edges, _ = plt.hist(arr, bins=bins, align='mid', edgecolor='black')

# Set x-ticks at the center of each bar
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])  # Midpoints of bins

# Integer labels

plt.xticks(bin_centers, labels=np.unique(arr))

plt.show()
```
