Create random data drawn from normal distribution

* size-how many data points to create
* loc-mean
* scale-std

In the histogram set density=True to normalize the values so thast the charges will be on the same scale.

```python

from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt

size = 1000
loc = 0
scale = 1


x= stats.norm.rvs(loc=loc, scale=scale, size=size)

y=norm.pdf(x, loc=loc, scale=scale)

plt.hist(x, density=True)

plt.plot(x, y)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 3
std = 5
size = 100

random = norm.rvs(loc=mean, scale=std, size=size)

x = np.linspace(mean - 4*std, mean + 4*std, 1000)

y = norm.pdf(x, loc=mean, scale=std)

plt.hist(random, bins=30, density=True, alpha=0.6, color='blue')

plt.plot(x, y, 'r-', lw=3)

plt.show()
```
