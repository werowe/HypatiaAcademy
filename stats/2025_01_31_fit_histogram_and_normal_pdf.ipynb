Create random data drawn from normal distribution

* size-how many data points to create
* loc-mean
* scale-std

In the histogram set density=True to normalize the values so thast the charges will be on the same scale.

```python

from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt

size = 100
loc = 0
scale = 1


x= stats.norm.rvs(loc=loc, scale=scale, size=size)

y=norm.pdf(x, loc=loc, scale=scale)

plt.hist(x, density=True)

plt.plot(x, y)


plt.show()
```
