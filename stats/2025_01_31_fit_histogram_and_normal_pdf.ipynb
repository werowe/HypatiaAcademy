```python
#PLOT THE HISTOGRAM


#import the libraries
from scipy.stats import norm,shapiro
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#specify the path
path = '/root/.cache/kagglehub/datasets/peimandaii/dataset-of-people/versions/1/hight.csv'

#fetch the data
data = pd.read_csv(path)

#print the first couple rows
print(data.head())


#get the data from the "Height" column
height_data = data['Hight']

#plot the histogram: specify the amount of bins, the color and density
plt.hist(height_data, bins=15, color='purple', density=True)

#name the grapgh
plt.title('Height distribution')

#plt.show
```

```python
#PLOTTING THE NORMAL DISTRIBUTION CURVE


#give values to mean and std
mean = 0
std = 1

#generate an array of evenly spaced numbers over a specified range
x = np.linspace(-4, 4, 100)

#calculate the probability density function
y = norm.pdf(x, mean, std)

#plot the graph
plt.plot(x, y)

#name it
plt.title("Normal Distribution Curve")

plt.show()
```

```python
#MERGE THE TWO


#plot the height prapgh again
plt.hist(height_data, bins=15, color='purple', density=True)

#get the mean and std values
mean_height = height_data.mean()
std_height = height_data.std()

#generate an array of 100 evenly spaced values ranging from min to max height
x = np.linspace(min(height_data), max(height_data), 100)

#calculate the probability density function
y = norm.pdf(x, mean_height, std_height)

#plot the graph
plt.plot(x,y, color = "yellow")


#CHECK IF IT IS THE NORMAL DiSTRIBUTION
stat, p_value = shapiro(height_data)
print(f"Shapiro-Wilk Test Stat: {stat}, P-value: {p_value}")

if p_value > 0.05:
    print("The data is a normal distribution")
else:
    print("The data is not a normal distribution")
```

Theory: People's height looks like a normal distribution when ploted

Proof:

1.   Plot a histogram using the data from a dataset
2.   Plot a normal distribution curve
3.   Put one on top of the other and check if they match using the Shapiro method

Conclusion: the distribution of human's height is roughly the same as a normal distribution
