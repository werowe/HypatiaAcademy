* size-how many data points to create
* loc-mean
* scale-std

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 3
std = 5
size = 100

random = norm.rvs(loc=mean, scale=std, size=size)

x = np.linspace(mean - 4*std, mean + 4*std, 1000)

y = norm.pdf(x, loc=mean, scale=std)

plt.hist(random, bins=30, density=True, alpha=0.6, color='blue')

plt.plot(x, y, 'r-', lw=3)

plt.show()
```

```python

stat, p_value = shapiro(x)
print(f"Shapiro-Wilk test stat: {stat}, p-value: {p_value}")

if p_value > 0.05:
    print("The data is a normal distribution")
else:
    print("The data is not a normal distribution")
```
