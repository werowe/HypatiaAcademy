# Definitions

* **entropy** measures impurity.  For example when all examples in the set are 1 then entropy is 100% and impurity is 0%.


* $P_1$ fraction of examples that are 1

* $p_0 = 1 - p_1$

**Entropy**:

$$H(p_1)$ = -p1 log_2(p_1) - (1 - P_1) log_2(p_1)$$


**Weighted Entropy**

Now we need another function to compute the weighted entropy in the splitted nodes. As you've seen in the video lecture, we must find:

- $w^{\text{left}}$ and $w^{\text{right}}$, the proportion of animals in **each node**.
- $p^{\text{left}}$ and $p^{\text{right}}$, the proportion of cats in **each split**.

Note the difference between these two definitions!! To illustrate, if we split the root node on the feature of index 0 (Ear Shape), then in the left node, the one that has the animals 0, 3, 4, 5 and 7, we have:

$$w^{\text{left}}= \frac{5}{10} = 0.5 \text{ and } p^{\text{left}} = \frac{4}{5}$$
$$w^{\text{right}}= \frac{5}{10} = 0.5 \text{ and } p^{\text{right}} = \frac{1}{5}$$

```python
def binary_entropy(p):
    # Handle edge cases (p=0 or p=1) to avoid log(0)
    if p == 0 or p == 1:
        return 0.0
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)
```

# Definitions

* $p_1^{left}$ is fraction in left node which are positive
* $w_1^{left}$ fraction of examples that went to the left node
* $p_1^{root}$ is fraction of example in the root node which are positive

* **information gain** is the reduction in entropy resulting from a split

$$H(p_1^{root}) = (w^{left} H(p_1^{left}) + w^{right} H(p_1^{right}))$$

Pick the feature with the highest information gain, will result in increasing the purity.

## Decision Tree Learning

- Start with all examples at the root node
- Calculate information gain for all possible features, and pick the one with the highest information gain
- Split dataset according to selected feature, and create left and right branches of the tree
- Keep repeating splitting process until stopping criteria is met:
  - When a node is 100% one class
  - When splitting a node will result in the tree exceeding a maximum depth
  - Information gain from additional splits is less than threshold
  - When number of examples in a node is below a threshold



## Reduce Overfitting


To reduce overfitting in a decision tree, several regularization and tuning methods can be applied to control the model’s complexity and improve generalization.​

1. Limit Tree Depth
Set a maximum depth (max_depth parameter) to prevent the tree from growing excessively complex.​

Shallow trees generalize better because they capture major trends rather than noise in the data.

2. Minimum Samples for Splitting and Leaves
Use parameters like min_samples_split and min_samples_leaf to require a minimum number of samples at a node before a split is made or a leaf is formed.​

Larger minimums mean fewer splits, resulting in simpler and more generalizable trees.

3. Pruning
Implement cost complexity pruning (or post-pruning), which removes branches that have little effect on the prediction error, simplifying the model.​

The parameter ccp_alpha in scikit-learn controls the pruning level: higher values prune more aggressively.

4. Minimum Impurity Decrease
Set a threshold (min_impurity_decrease) for the impurity reduction required to create a new split.​

This prevents splits that contribute little to predictive power, reducing the risk of fitting random variation.

5. Data Regularization and Validation
Clean and balance the dataset to avoid biased splits on noise or small subgroups.​

Use cross-validation to evaluate model performance on multiple subsets and monitor generalization errors.​

6. Ensemble Techniques
Use methods like Random Forests or Gradient Boosted Trees, which combine multiple weak learners (shallow trees) to improve generalization and reduce variance.​

Bagging reduces variance by averaging multiple trees trained on bootstrap samples, while boosting sequentially corrects errors from prior trees.

7. Early Stopping (for Boosted Trees)
In algorithms such as XGBoost or LightGBM, early stopping halts training when validation loss stops improving—preventing overfitting through excessive learning cycles.​
