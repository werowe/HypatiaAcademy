```python
import pandas as pd

from datetime import date

df = pd.DataFrame({
   'sales' : [100,200,300,400,500],
    'date': [date.fromisoformat("2025-02-22"),
             date.fromisoformat("2025-02-22"),
             date.fromisoformat("2025-02-24"),
             date.fromisoformat("2025-02-25"),
             date.fromisoformat("2025-02-26")]

})



df['date']=pd.to_datetime(df['date'])

df.set_index('date', inplace=True)

df['sales'].plot(kind='bar')
```
