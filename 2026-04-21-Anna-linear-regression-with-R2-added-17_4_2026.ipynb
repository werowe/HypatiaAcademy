## What is R² (R-squared) in linear regression?

**R² measures how well your regression model explains the variation in the data.**

---

### 1) The idea (intuitive)

Think of your data points scattered around. Your regression line tries to “fit” them.

- **R² = 1** → perfect fit (all points lie exactly on the line)  
- **R² = 0** → model explains nothing (no better than predicting the average)  
- **R² = 0.7** → model explains 70% of the variation in the target  

---

### 2) The formula

$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$

Where:
- $y_i$ = actual values  
- $\hat{y}_i$ = predicted values  
- $\bar{y}$ = mean of the data  

---

### 3) What it’s really comparing

R² compares two things:

1. **Your model’s error**  
   $
   \sum (y_i - \hat{y}_i)^2
   $

2. **A baseline model (predict the mean)**  
   $
   \sum (y_i - \bar{y})^2
   $

👉 So:
- If your model is much better than the mean → R² close to 1  
- If it’s no better → R² ≈ 0  
- If it’s worse → **R² can be negative**

---

### 4) Quick example

Suppose you're predicting house prices:

- Mean-only model → large errors  
- Linear regression → smaller errors  

If regression reduces error by 80% → **R² = 0.80**

---

### 5) Important caveats

- **High R² ≠ good model**
  - Could be overfitting  
- **Low R² ≠ useless**
  - Some problems are inherently noisy  
- **Adding variables always increases R²**
  - That’s why *adjusted R²* is often used  

---

### 6) Intuition (ML perspective)

You can think of R² as:

> “How much variance did my model capture compared to just predicting a constant?”
