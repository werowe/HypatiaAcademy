```python

import numpy as np

import matplotlib.pyplot as plt


arr = np.random.randint(1,5,20)

print("arr sorted", np.sort(arr))

unique_values, counts = np.unique(arr, return_counts=True)

labels = np.unique(arr)

plt.pie(counts, labels=labels, colors=["yellow", "blue", "green", "red"], autopct="%1.1f%%")

plt.legend(title="Values")

plt.show()
```

```python
# add explode
import numpy as np

import matplotlib.pyplot as plt



unique_values, counts = np.unique(arr, return_counts=True)

labels = np.unique(arr)

explode=[0,0.1,0,0]

plt.pie(counts, labels=labels, explode=explode, colors=["yellow", "blue", "green", "red"], autopct="%1.1f%%")

plt.legend(title="Values")

plt.show()
```
