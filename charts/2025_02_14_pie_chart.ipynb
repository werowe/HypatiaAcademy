```python

import numpy as np

import matplotlib.pyplot as plt


arr = np.random.randint(1,5,20)

unique_values, counts = np.unique(arr, return_counts=True)

labels = np.unique(arr)

plt.pie(counts, labels=labels, colors=["yellow", "blue", "green", "red"], autopct="%1.1f%%")

plt.legend(title="Values")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


sizes = [35, 25, 25, 15]
labels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
explode = [0.2,0,0,0]


plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title("Exploded Pie Chart Example")
plt.show()
```
