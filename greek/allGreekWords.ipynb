```python
# Define a dictionary for Greek to phonetic Latin transliteration
transliteration_dict = {
    'Α': 'A', 'α': 'a', 'Β': 'V', 'β': 'v',
    'Γ': 'G', 'γ': 'g', 'Δ': 'D', 'δ': 'd',
    'Ε': 'E', 'ε': 'e', 'Ζ': 'Z', 'ζ': 'z',
    'Η': 'I', 'η': 'i', 'Θ': 'Th', 'θ': 'th',
    'Ι': 'I', 'ι': 'i', 'Κ': 'K', 'κ': 'k',
    'Λ': 'L', 'λ': 'l', 'Μ': 'M', 'μ': 'm',
    'Ν': 'N', 'ν': 'n', 'Ξ': 'X', 'ξ': 'x',
    'Ο': 'O', 'ο': 'o', 'Π': 'P', 'π': 'p',
    'Ρ': 'R', 'ρ': 'r', 'Σ': 'S', 'σ': 's', 'ς': 's',
    'Τ': 'T', 'τ': 't', 'Υ': 'Y', 'υ': 'y',
    'Φ': 'F', 'φ': 'f', 'Χ': 'Ch', 'χ': 'ch',
    'Ψ': 'Ps', 'ψ': 'ps', 'Ω': 'O', 'ω': 'o'
}

def transliterate_greek(word):
    return ''.join(transliteration_dict.get(char, char) for char in word)

 

 
```
