# Video 2: Cómo aprende una red neuronal

Este notebook acompaña el [segundo video](https://youtu.be/iFlQrSNQ7pQ) de la serie sobre redes neuronales en PyTorch.

En el primer video vimos el **paso hacia adelante**: cómo una red neuronal toma datos de entrada, los combina con pesos y un sesgo, y produce una predicción. En ese momento, el modelo todavía no estaba aprendiendo. Solo estaba calculando una salida.

En este segundo video damos el siguiente paso: vemos **cómo aprende una red neuronal**.

La idea principal es que el modelo hace una predicción, compara esa predicción con la respuesta correcta, calcula una pérdida, y luego usa los gradientes para ajustar los pesos.

El ciclo completo es:

```text
predicción → pérdida → gradientes → actualización de pesos → repetir

````python

import torch
import torch.nn.functional as F

# Fijamos la semilla para que los resultados sean reproducibles
torch.manual_seed(42)

# Creamos 100 ejemplos, cada uno con 5 características
X = torch.rand(100, 5)

# Estos son los pesos "reales" que usamos para crear las respuestas
# En un problema real, normalmente no conocemos estos pesos
true_weights = torch.tensor(
    [[2.0], [1.5], [1.0], [2.5], [1.0]]
)

# Calculamos una puntuación para cada fila de datos
score = X @ true_weights

# Creamos las etiquetas reales:
# si la puntuación es mayor que el promedio, la clase es 1
# si no, la clase es 0
y = (score > score.mean()).float()

# Creamos los pesos del modelo de forma aleatoria
# requires_grad=True permite que PyTorch calcule los gradientes
W = torch.randn(5, 1, requires_grad=True)

# Creamos el sesgo del modelo
b = torch.zeros(1, requires_grad=True)

# La tasa de aprendizaje controla el tamaño de cada paso
learning_rate = 0.1

# Repetimos el proceso de entrenamiento 100 veces
for epoch in range(100):

    # Paso hacia adelante:
    # combinamos los datos con los pesos y el sesgo
    z = X @ W + b

    # Aplicamos la función sigmoide para obtener predicciones entre 0 y 1
    y_hat = torch.sigmoid(z)

    # Calculamos la pérdida comparando la predicción con la respuesta real
    loss = F.binary_cross_entropy(y_hat, y)

    # Paso hacia atrás:
    # PyTorch calcula los gradientes de la pérdida
    loss.backward()

    # Actualizamos los pesos sin registrar esta operación en el grafo
    with torch.no_grad():

        # Movemos los pesos en la dirección contraria al gradiente
        W -= learning_rate * W.grad

        # Hacemos lo mismo con el sesgo
        b -= learning_rate * b.grad

        # Limpiamos los gradientes para la siguiente época
        W.grad.zero_()
        b.grad.zero_()

    # Cada 10 épocas mostramos la pérdida
    if epoch % 10 == 0:
        print(f"epoch {epoch}, "
              f"loss = {loss.item():.4f}")

````
