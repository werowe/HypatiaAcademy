# Entrenamiento completo de una red neuronal simple en PyTorch

En este notebook usamos la versión estándar de PyTorch para entrenar un modelo simple de clasificación binaria.

En los ejemplos anteriores calculamos varias cosas manualmente para entender mejor las ideas: el paso hacia adelante, la pérdida, los gradientes y la actualización de los pesos.

Aquí hacemos la misma idea, pero usando las herramientas normales de PyTorch.

Primero creamos un conjunto pequeño de datos con 100 ejemplos. Cada ejemplo tiene 5 características. Después creamos etiquetas binarias: `0` o `1`.

Luego definimos el modelo con:

```python
nn.Linear(5, 1)

````python
import torch
import torch.nn as nn

# Fijamos la semilla para que los números aleatorios sean reproducibles.
# Así, cada vez que ejecutamos el notebook, obtenemos los mismos datos iniciales.
torch.manual_seed(42)

# Creamos 100 ejemplos de entrenamiento.
# Cada ejemplo tiene 5 características.
X = torch.rand(100, 5)

# Estos son los pesos "verdaderos" que usamos para crear los datos.
# En un problema real no conoceríamos estos pesos.
true_weights = torch.tensor(
    [[2.0], [1.5], [1.0], [2.5], [1.0]]
)

# Calculamos una puntuación para cada fila de X.
# Esto simula una relación real entre las características y la salida.
score = X @ true_weights

# Creamos las etiquetas y.
# Si la puntuación está por encima del promedio, la etiqueta es 1.
# Si no, la etiqueta es 0.
# Esto convierte el problema en una clasificación binaria.
y = (score > score.mean()).float()

# Creamos el modelo.
# nn.Linear(5, 1) significa:
# 5 valores de entrada y 1 valor de salida.
# PyTorch crea automáticamente los pesos y el sesgo.
model = nn.Linear(5, 1)

# Definimos la función de pérdida.
# BCEWithLogitsLoss combina la sigmoide y la entropía cruzada binaria.
# Por eso el modelo produce logits, no probabilidades directamente.
loss_fn = nn.BCEWithLogitsLoss()

# Definimos el optimizador.
# SGD actualiza los pesos y el sesgo del modelo.
# model.parameters() le dice al optimizador qué parámetros debe actualizar.
# lr es la tasa de aprendizaje.
optimizer = torch.optim.SGD(
    model.parameters(), lr=0.1
)

# Entrenamos el modelo durante 100 épocas.
for epoch in range(100):

    # Paso hacia adelante:
    # el modelo recibe X y produce logits.
    logits = model(X)

    # Calculamos la pérdida comparando los logits con las etiquetas reales.
    loss = loss_fn(logits, y)

    # Borramos los gradientes anteriores.
    # PyTorch acumula gradientes, así que hay que limpiarlos antes de cada paso.
    optimizer.zero_grad()

    # Calculamos los nuevos gradientes con backpropagation.
    loss.backward()

    # Actualizamos los pesos y el sesgo usando los gradientes.
    optimizer.step()

    # Mostramos la pérdida cada 10 épocas.
    if epoch % 10 == 0:
        print(f"epoch {epoch}, "
              f"loss = {loss.item():.4f}")
````
