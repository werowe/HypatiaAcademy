```python
import torch
import torch.nn.functional as F

torch.manual_seed(42)

X = torch.rand(100, 5)

# hidden rule:

# students pass if a weighted combination of features is high enough

true_weights = torch.tensor([[2.0], [1.5], [1.0], [2.5], [1.0]])

score = X @ true_weights

y = (score > score.mean()).float()

W = torch.randn(5, 1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

learning_rate = 0.1

for epoch in range(100):
    z = X @ W + b
    y_hat = torch.sigmoid(z)

    loss = F.binary_cross_entropy(y_hat, y)

    loss.backward()

    with torch.no_grad():
        W -= learning_rate * W.grad
        b -= learning_rate * b.grad

        W.grad.zero_()
        b.grad.zero_()

    if epoch % 10 == 0:
        print(f"epoch {epoch}, loss = {loss.item():.4f}")

```

### Manual Calculation of Loss and Gradients

Let's manually calculate the binary cross-entropy loss and its gradients with respect to `W` and `b` for a single step.

```python
import torch
import torch.nn.functional as F

# Re-initialize weights and bias for demonstration
W_manual = torch.randn(5, 1, requires_grad=True)
b_manual = torch.zeros(1, requires_grad=True)

# Perform one forward pass to get y_hat
z_manual = X @ W_manual + b_manual
y_hat_manual = torch.sigmoid(z_manual)

# 1. Manual Loss Calculation
# BCE Loss: -(y * log(y_hat) + (1 - y) * log(1 - y_hat))
# torch.log(y_hat_manual) will be unstable for values close to 0 or 1.
# F.binary_cross_entropy handles this stability internally.
# For manual calculation, we'll use a clipped version to avoid log(0).
epsilon = 1e-7 # Small value to prevent log(0)
y_hat_clipped = torch.clamp(y_hat_manual, epsilon, 1 - epsilon)

manual_loss = -torch.mean(y * torch.log(y_hat_clipped) + (1 - y) * torch.log(1 - y_hat_clipped))
print(f"Manual BCE Loss: {manual_loss.item():.4f}")

# Compare with F.binary_cross_entropy for the same y_hat and y
builtin_loss = F.binary_cross_entropy(y_hat_manual, y)
print(f"Built-in BCE Loss: {builtin_loss.item():.4f}")

# 2. Manual Gradient Calculation
# Gradient of BCE loss with respect to z (logits) is (y_hat - y)
# dLoss/dz = y_hat - y
grad_z = y_hat_manual - y

# z = X @ W + b
# dLoss/dW = X.T @ dLoss/dz
grad_W_manual = X.T @ grad_z

# dLoss/db = sum(dLoss/dz)
grad_b_manual = grad_z.sum()

print("\nManual Gradients:")
print(f"Gradient of W (first 5 elements): {grad_W_manual[:5].flatten().tolist()}")
print(f"Gradient of b: {grad_b_manual.item():.4f}")

# To verify, let's use autograd for comparison
manual_loss.backward()
print("\nAutograd Gradients:")
print(f"Gradient of W (first 5 elements): {W_manual.grad[:5].flatten().tolist()}")
print(f"Gradient of b: {b_manual.grad.item():.4f}")
```

# Then add more layers

```python
import torch
import torch.nn as nn

torch.manual_seed(42)

X = torch.rand(100, 5)

# hidden rule:

# students pass if a weighted combination of features is high enough

true_weights = torch.tensor([[2.0], [1.5], [1.0], [2.5], [1.0]])

score = X @ true_weights

y = (score > score.mean()).float()

model = nn.Sequential(
    nn.Linear(5, 8),
    nn.ReLU(),
    nn.Linear(8, 1)
)

loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(100):
    logits = model(X)
    loss = loss_fn(logits, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"epoch {epoch}, loss = {loss.item():.4f}")
```

# Then predictions

```python
with torch.no_grad():
    # Desactiva el cálculo de gradientes. Esto es importante para la inferencia
    # ya que no necesitamos calcular gradientes para actualizar los pesos del modelo.
    # Mejora la eficiencia computacional y reduce el uso de memoria.

    logits = model(X)
    # Pasa los datos de entrada 'X' a través del modelo (red neuronal) para obtener
    # las 'logits'. Las logits son las salidas crudas del modelo antes de aplicar
    # una función de activación para convertir en probabilidades.

    probabilities = torch.sigmoid(logits)
    # Aplica la función sigmoide a las logits. La función sigmoide comprime los valores
    # de entrada entre 0 y 1, lo que los convierte en probabilidades interpretables.
    # En problemas de clasificación binaria, esto representa la probabilidad de la clase positiva.

    predictions = (probabilities >= 0.5).float()
    # Convierte las probabilidades en predicciones binarias. Si la probabilidad es
    # mayor o igual a 0.5, la predicción se considera 1 (clase positiva); de lo contrario, es 0.
    # El método `.float()` asegura que las predicciones sean de tipo flotante (0.0 o 1.0).

accuracy = (predictions == y).float().mean()
# Calcula la precisión del modelo. Compara las predicciones ('predictions') con las
# etiquetas verdaderas ('y').
# `(predictions == y)` crea un tensor booleano donde `True` indica una predicción correcta.
# `.float()` convierte estos booleanos a 1.0 (True) o 0.0 (False).
# `.mean()` calcula el promedio de estos valores, lo que da la proporción de predicciones correctas.

print("accuracy =", accuracy.item())
# Imprime la precisión calculada. `.item()` extrae el valor escalar del tensor de un solo elemento.
```
